```python
#Line Plot
import matplotlib.pyplot as pp
pp.plot([1,3,5,6,7], [1, 2, 3, 4, 6])
pp.axis([0, 7, 0, 10])
pp.show()
```

```python
pp.bar([0.25,2.25,5.25],[300,200,600],
label="Marks",color='b',width=0.5)
pp.bar([0.75,2.75,5.75],[100,120,150],
label="Previous Marks", color='g',width=.5)
pp.legend()
pp.xlabel('Days')
pp.ylabel('Wage')
pp.title('Details')
pp.show()
```

```python
#Histrogram Plot
import matplotlib.pyplot as plt
X = [22,55,62,45,21,22,34,34,55,77,22,84,4,7,20,10,3,2]
Y = [1,10,20,30,40,50]
plt.hist(X, Y, rwidth=0.6)
plt.xlabel('age groups')
plt.ylabel('Number of people')
plt.title('Histogram')
plt.show()
```

```python
# Box Plot-They are used to show distributions of numeric data values,especially wh
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(10)
data = np.random.normal(100, 20, 200)
fig = plt.figure(figsize =(10, 7))
plt.boxplot(data)
plt.show()
```
